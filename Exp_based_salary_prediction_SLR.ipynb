```python
import pandas as pd #importing Pandas library to perform data and file operations
```

```python
dataset=pd.read_csv("Salary_Data.csv") #loading data from CSV file to dataset variable
```

```python
dataset
```

```python
ind_var=dataset[["YearsExperience"]]#assigning the data from yearsExperience(independent variable/input) column to variable
dep_var=dataset[["Salary"]]#assigning the data from Salary(dependent variable/output) column to variable

ind_var
```

```python
dep_var
```

```python
import matplotlib.pyplot as plt 
#importing matplotlib library to perform chart operations
```

```python
plt.scatter(ind_var,dep_var)#calling scatter function to mark the datapoints based on inputs
plt.xlabel("experience")
plt.ylabel("salary")
plt.show()
```

```python
from sklearn.model_selection import train_test_split #importing scikit library to perform ML operations
X_train,X_test,Y_train,Y_test=train_test_split(ind_var,dep_var,test_size=0.30,random_state=0)
#calling the function to split train and test set and assigning their output to respective training and testing variables
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression #importing LinearRegression class to train the model using SLR algorithm
regressor=LinearRegression()#creating instance for the class
regressor.fit(X_train,Y_train)#training the model by giving training set as input
```

```python
regressor.coef_# this is the weight in y=wx+b
```

```python
regressor.intercept_#this the bias in y=wx+b
```

```python
Y_pred=regressor.predict(X_test)#running the trained model over test set and getting predicted output
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import r2_score#importing the metrics library to do evaluations operations
r_score=r2_score(Y_test,Y_pred)#finding the R^2 (R square)
```

```python
r_score#if R2 is nearer to 1 then model is good
```

```python
import pickle as pk
filename="SLR_model.sav"
pk.dump(regressor,open(filename,'wb'))#savinf the model under the given filename
```

```python
saved_model=pk.load(open(filename,'rb'))#loading the model
```

```python
salary=saved_model.predict([[15]])#predicting the salary using the trained model
```
